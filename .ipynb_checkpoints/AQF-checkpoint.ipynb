```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
HL = 2
HU = 64
SMV = 4
CSMV = 32
CHL= 0
CHU = 8
LR = 0.001
DR = 0.05
ACT = 'relu'
```

## Example 2

```python
np.random.seed(1)

N_assets = 500
n_asset_index = 1
n_macro_index = 2
T_train = 250
T_valid = 100
T_test = 250
T = T_train + T_valid + T_test

n_macro_info_State = SMV
n_asset_info_State = n_asset_index + n_macro_info_State
n_macro_info_Moment = CSMV  # Moment output dimension
n_moment = CHU

def generator1(T_train,T_valid,T_test,N_assets,n_asset_index, n_macro_index,seed):
    np.random.seed(seed)     
    T = T_train + T_valid + T_test
    T0, T1, T2, T3 = 0, T_train, T_train + T_valid, T
    asset_index = np.random.normal(size = (1, T, N_assets,n_asset_index))
    C1 = asset_index[:,:,:,0]
    
    
    dates = np.arange(T)
    h = np.sin(dates/24*np.pi) + np.random.normal(0,0.5,T)
    bh0 = 2*(h>0)-1.0
    bh = np.reshape(bh0,(1,T,1))
    bh = np.tile(bh,(1,1,N_assets))
    C2 = bh
    
    beta = C1*C2
    
    macro_observed = 0.05*dates + h
    macro_index0 = np.reshape(macro_observed,(1,T,1))
    D = np.reshape(dates,(1,T,1))
#     macro_index = macro_index0
    macro_index = np.concatenate((macro_index0,D),axis = -1)
#     macro_index = np.reshape(h,(1,T,1))
    
    
    sigma_F = 0.1
    mu_F = np.sqrt(sigma_F)
    F0 = np.random.normal(mu_F,np.sqrt(sigma_F),size = (1,T,1))
    F = np.tile(F0,(1, 1, N_assets))
    
    mu_epsilon = 0
    sigma_epsilon = 1
    epsilon = np.random.normal(mu_epsilon, sigma_epsilon, size = (1, T,N_assets))
    
    R = beta*F + epsilon
    R = np.reshape(R,[1,T,N_assets,1])
    print('macro_index  ','asset_index  ','R  ')
    print(macro_index.shape, asset_index.shape, R.shape)

    cost_discount = np.ones(shape = (1,T,N_assets,n_moment))
    
    x = [macro_index, asset_index, R, cost_discount]
    y = np.zeros(shape = (T, N_assets,1))
    
    x_train = [macro_index[:,0:T_train,:], asset_index[:,0:T_train,:,:], R[:,0:T_train,:,:], cost_discount[:,0:T_train,:,:]]
    y_train = np.zeros(shape = (1,T_train, N_assets,1))
    F_train = F0[:,0:T_train,:]
    
    x_valid = [macro_index[:,T1:T2,:], asset_index[:,T1:T2,:,:], R[:,T1:T2,:,:], cost_discount[:,T1:T2,:,:]]
    y_valid = np.zeros(shape = (1,T_valid, N_assets,1))
    F_valid = F0[:,T_valid,:]
    
    x_test = [macro_index[:,-T_test:,:], asset_index[:,-T_test:,:,:], R[:,-T_test:,:,:],cost_discount[:,-T_test:,:,:]]
    y_test = np.zeros(shape = (1,T_test, N_assets,1))
    F_test = F0[:,-T_test:,:]
    
    plt.rcParams['figure.figsize'] = (12.0, 4.0)
    plt.title("Observed Macro Index")
    plt.scatter(dates[:T_train],macro_observed[:T_train], label='Train')
    plt.scatter(dates[T1:T2],macro_observed[T1:T2], label='Validation')
    plt.scatter(dates[-T_test:],macro_observed[-T_test:], label='Test')
    plt.legend(loc = 2)
    plt.grid()
    plt.show()
    plt.title("Hidden Macro Index")
    plt.scatter(dates[:T_train],h[:T_train], label='Train')
    plt.scatter(dates[T1:T2],h[T1:T2], label='Validation')
    plt.scatter(dates[-T_test:],h[-T_test:], label='Test')
    plt.legend(loc = 2)
    plt.grid()
    plt.show()
    plt.title("Characteristic Index")
    plt.scatter(dates[:T_train],bh0[:T_train], label='Train')
    plt.scatter(dates[T1:T2],bh0[T1:T2], label='Validation')
    plt.scatter(dates[-T_test:],bh0[-T_test:], label='Test')
    plt.legend(loc = 2)
    plt.grid()
    plt.show()
    
    return x, y, F0, x_train, y_train, F_train, x_valid, y_valid, F_valid,\
    x_test, y_test, F_test, dates, h, beta, epsilon

x, y, F, x_train, y_train, F_train, x_valid, y_valid, F_valid, x_test, y_test, F_test,\
dates, h, beta, ep = generator1(T_train,T_valid,T_test,N_assets,n_asset_index, n_macro_index,1)
```

```python
tf.random.set_seed(0)

# State RNN       
# (Batch, T, n_macro_index) --------> (Batch, T, n_macro_info)
inputs = keras.Input(shape=(None, n_macro_index)) # 
# State_FFN0 = layers.TimeDistributed(layers.Dense(1, activation = 'linear'))
State_RNN1 = layers.LSTM(n_macro_info_State,return_sequences = True, dropout = 0.05)
# outputs = State_FFN0(inputs)
outputs = State_RNN1(inputs)
State_RNN = keras.Model(inputs=inputs, outputs=outputs, name = 'State_RNN')
State_RNN.summary()
```

```python
# State FFN model
# n_asset_info = n_asset_index + n_macro_info
# (Batch, n_asset_info) --------> (Batch, w)

n_asset_info_State = n_asset_index + n_macro_info_State
inputs = keras.Input(shape = (n_asset_info_State))
outputs = inputs

State_FFN = [0] * HL
db = [0] * HL
for i in range(HL):
    State_FFN_tmp = layers.Dense(HU, activation = ACT)
    db_tmp = layers.Dropout(0.05)
    db1 = layers.Dropout(0.05)
    State_FFN[i] = State_FFN_tmp
    db[i] = db_tmp 
State_FFN0 = layers.Dense(1, activation = 'linear')  
db0 = layers.Dropout(0.05)
for i in range(HL):
    outputs = State_FFN[i](outputs)
    outputs = db[i](outputs)
outputs = State_FFN0(outputs)
outputs = db0(outputs)
State_FFN = keras.Model(inputs=inputs, outputs=outputs, name = 'State_FFN')
State_FFN.summary()

# Global State FFN model
# (Batch, T, N_assets, n_asset_info) --------> (Batch, T, N_assets, w)
inputs = keras.Input(shape = (None, N_assets, n_asset_info_State))    # asset index
TD_State_FFN = layers.TimeDistributed(layers.TimeDistributed(State_FFN))
outputs = TD_State_FFN(inputs)
Global_State_FFN = keras.Model(inputs=inputs, outputs=outputs, name = 'Global_State_FFN')
# Global_State_FFN.summary()
```

```python
# State  
# [ (Batch, T, n_macro_index), (Batch, T, N_assets, n_asset_index),  (Batch, T, N_assets)]


macro_index = keras.Input(shape=(None, n_macro_index))
asset_index = keras.Input(shape=(None, N_assets, n_asset_index))
R = keras.Input(shape=(None, N_assets,1))
inputs = [macro_index, asset_index, R]

macro_info = State_RNN(macro_index) 
batch_tmp, time_tmp = tf.shape(macro_info)[0], tf.shape(macro_info)[1]
macro_info0 = tf.reshape(macro_info,[batch_tmp, time_tmp, 1, n_macro_info_State])
macro_info_tiled = tf.tile(macro_info0,[1,1,N_assets,1])

asset_info = tf.concat([macro_info_tiled, asset_index],axis = -1)
w = Global_State_FFN(asset_info)
M = tf.multiply(R,w)
M = tf.math.reduce_sum(M, axis = [3,2],keepdims=True)
M = 1 - M
print(M.shape)
State = keras.Model(inputs=inputs, outputs = [M,w,asset_info], name = 'State')   # output the M on each day
State.summary()
# output has shape (time_steps, 1)
```

```python
# Moment RNN
# (Batch, T, n_macro_index) --------> (Batch, T, n_macro_info)
n_macro_info_Moment = CSMV  # Moment output dimension
inputs = keras.Input(shape=(None, n_macro_index))
Moment_RNN1 = layers.LSTM(n_macro_info_Moment,return_sequences = True, dropout = 0.05)
outputs = Moment_RNN1(inputs)
Moment_RNN = keras.Model(inputs=inputs, outputs=outputs, name = 'Moment_RNN')
# Moment_RNN.summary()
```

```python
# Moment FFN model
# (Batch, n_asset_info) --------> (Batch, n_moment)
n_moment = CHU
n_asset_info_Moment = n_asset_index + n_macro_info_Moment
inputs = keras.Input(shape = (n_asset_info_Moment))
Moment_FFN = layers.Dense(n_moment, activation = 'tanh')
db = layers.Dropout(0.5)
if CHL == 0:    
    outputs = Moment_FFN(inputs)
    outputs = db(outputs)
else:
    Moment_FFN0 = layers.Dense(n_moment, activation = ACT)
    db0 = layers.Dropout(0.5)
    outputs = Moment_FFN0(inputs)
    outputs = db0(outputs)
    outputs = Moment_FFN(outputs)
    outputs = db(outputs)
Moment_FFN = keras.Model(inputs=inputs, outputs=outputs, name = 'Moment_FFN')
# Moment_FFN.summary()

# Global Moment FFN model
# (Batch, T, N_assets, n_asset_info) --------> (Batch, T, N_assets, n_moment)
inputs = keras.Input(shape = (None, N_assets, n_asset_info_Moment))    # asset index
TD_Moment_FFN = layers.TimeDistributed(layers.TimeDistributed(Moment_FFN))
outputs = TD_Moment_FFN(inputs)
Global_Moment_FFN = keras.Model(inputs=inputs, outputs=outputs, name = 'Global_Moment_FFN')
# Global_Moment_FFN.summary()
```

```python
# Moment
macro_index = keras.Input(shape=(None,n_macro_index))
asset_index = keras.Input(shape=(None,N_assets, n_asset_index))
inputs = [macro_index, asset_index]
macro_info = Moment_RNN(macro_index)

batch_tmp, time_tmp = tf.shape(macro_info)[0], tf.shape(macro_info)[1]
macro_info0 = tf.reshape(macro_info,[batch_tmp, time_tmp, 1, n_macro_info_Moment])
macro_info_tiled = tf.tile(macro_info0,[1,1,N_assets,1])
asset_info = tf.concat([macro_info_tiled, asset_index],axis = -1)

g = Global_Moment_FFN(asset_info)
Moment = keras.Model(inputs=inputs, outputs=g, name = 'Moment')

# Moment.summary()
```

```python
# Combined 
macro_index = keras.Input(shape=(None, n_macro_index))
asset_index = keras.Input(shape=(None, N_assets, n_asset_index))
R = keras.Input(shape=(None, N_assets,1))
cost_dc = keras.Input(shape=(None, N_assets, n_moment))
print(macro_index.shape, asset_index.shape, R.shape, cost_dc.shape)
inputs = [macro_index, asset_index, R, cost_dc]

Mandw = State([macro_index, asset_index, R])
M = Mandw[0] # shape = (1,T)
g = Moment([macro_index, asset_index])
print(M.shape,R.shape,g.shape)
M_tiled = tf.tile(M,[1,1,N_assets,n_moment])
R_tiled = tf.tile(R,[1,1,1,n_moment])
print(M_tiled.shape,R_tiled.shape,g.shape)
outputs = tf.multiply(M_tiled,R_tiled)
outputs = tf.multiply(outputs,g)
outputs = tf.multiply(outputs,cost_dc)
Combined = keras.Model(inputs=inputs, outputs=outputs, name = 'Combined')
Combined.summary()
# output has shape (time_steps, 1)
# keras.utils.plot_model(Combined, 'my_first_model.png')
```

```python
def loss_State(y_true,y_pred):
    z = y_pred[0] - y_true[0]
    z = tf.reduce_sum(z,0)
    z = tf.math.square(z)
    z = tf.reduce_mean(z, [1,0])
    return z

def loss_Moment(y_true,y_pred):
    z = y_pred[0] - y_true[0]
    z = tf.reduce_sum(z,0)
    z = tf.math.square(z)
    z = tf.reduce_mean(z, [1,0])
    return -z
```

```python
# First Train
for l in Moment.layers:
    l.trainable = False
for l in State.layers:
    l.trainable = True
    
custom_optimizer = keras.optimizers.Adam(learning_rate = LR)
Combined.compile(optimizer=custom_optimizer,loss=loss_State)
Combined.fit(x=x_train,y=y_train, epochs=200,verbose=1)
```

```python
# Second Train
for l in Moment.layers:
    l.trainable = True
for l in State.layers:
    l.trainable = False
custom_optimizer = keras.optimizers.Adam(learning_rate = LR)
Combined.compile(optimizer = custom_optimizer,loss=loss_Moment)
Combined.fit(x=x_train,y=y_train, epochs=200,verbose=1)
```

```python
for l in Moment.layers:
    l.trainable = False
for l in State.layers:
    l.trainable = True
custom_optimizer = keras.optimizers.Adam(learning_rate = LR)
Combined.compile(optimizer = custom_optimizer,loss=loss_Moment)
Combined.fit(x=x_train,y=y_train, epochs=200,verbose=1)
```

```python
def aux_output(x):
    opt = State.predict(x)
    w = opt[1]
    hidden = opt[2]
    R = x[2]
    wR = w*R
    F = np.sum(wR[0],axis = 1)
    return w, hidden, F
def sharpe(F):
    return F.mean()/F.std()
```

```python
w_GAN, hidden, F_GAN = aux_output(x)
F_GAN_train = F_GAN[:T_train]
F_GAN_test= F_GAN[-T_test:]
```

```python
print('real sharpe train: ',F_train.mean()/F_train.std())
print('real sharpe test: ',F_test.mean()/F_test.std())
print('GAN sharpe train: ', sharpe(F_GAN_train))
print('GAN sharpe test: ', sharpe(F_GAN_test))
```

```python
inputs = keras.Input(shape = (None, n_asset_info_State))
State_FFN1 = layers.Dense(8, activation = 'relu')
State_FFN2 = layers.Dense(1, activation = 'linear')
db1 = layers.Dropout(0.5)
db2 = layers.Dropout(0.5)
outputs = layers.TimeDistributed(State_FFN1)(inputs)
outputs = layers.TimeDistributed(db1)(outputs)
outputs = layers.TimeDistributed(State_FFN2)(outputs)
outputs = layers.TimeDistributed(db2)(outputs)
beta_FFN = keras.Model(inputs=inputs, outputs=outputs, name = 'beta_FFN')
# beta_FFN.summary()
```

```python
def cl(y1,y2):
    z = y1-y2
    z = tf.math.square(z)
    z = tf.reduce_sum(z,axis = [2,1])
    z = tf.math.sqrt(z)
    z = tf.reduce_mean(z,axis = 0)
    return z
beta_FFN.compile(loss = cl, optimizer = 'Adam')
beta_FFN.fit(hidden_input, r*f, epochs = 100, verbose = 1)
```
